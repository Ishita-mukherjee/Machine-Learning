# Checking with linear regression

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

```python
# Importing the dataset
dataset = pd.read_csv(r'C:\Users\LENOVO\Ardent ML\Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:2].values # This line creates a matrix
y = dataset.iloc[:, 2].values 
```

```python
# Fitting Linear Regression to the dataset
lin_reg = LinearRegression()
lin_reg.fit(X, y) 
```

```python
salary = lin_reg.predict([[8.3]])

print("Salary for 8.3 Label:",salary)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X,y,color='orange') #original
plt.plot(X, lin_reg.predict(X), color = 'blue') #predicted

plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\Users\LENOVO\Ardent ML\Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:2].values # This line creates a matrix
y = dataset.iloc[:, 2].values
```

```python
from sklearn.preprocessing import PolynomialFeatures

#hyper parameter pruning or Bias value trade off
poly_reg = PolynomialFeatures(degree = 5) #hyper-parameter
X_poly = poly_reg.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Predicting a new result with Polynomial Regression
k=poly_reg.fit_transform([[8.3]])
salary = lin_reg_2.predict(k)
print("Predicted Salary:",salary)
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly),
         color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

```
