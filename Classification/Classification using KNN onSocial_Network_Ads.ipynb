```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Importing the dataset
dataset = pd.read_csv(r'C:\Users\LENOVO\Ardent ML\Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, [2, 3]].values
X
```

```python
y = dataset.iloc[:, 4].values
y
```

```python
# Splitting the dataset into the Training set and Test set
#from sklearn.cross_validation import train_test_split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```python
# Feature Scaling to decrease the range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) #on train we do fit_transform
X_test = sc.transform(X_test) #on test we do transform
```

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(weights='distance')
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
#Confusion matrix to find the number of matched and unmatched data as we cannot use rmse here as all values 0 or 1

from sklearn.metrics import confusion_matrix
print("My Confusion Matrix:",confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy: ",accuracy_score(y_test, y_pred))
```

```python
#new customer prediction age=28 salry=119000

k=classifier.predict(sc.transform([[28,119000]])) #need to scale this also else will get wrong output
print("New customer class:",k)
if k[0]==0:
    print("Not purchase type")
else:
    print("Purchase type")
```

```python
#plotting the graph
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
aranged_ages = np.arange(start = X_set[:, 0].min(),stop = X_set[:, 0].max(), step = 0.01)
aranged_salaries = np.arange(start = X_set[:, 1].min(), stop = X_set[:, 1].max(), step = 0.01)
```

```python
X1, X2 = np.meshgrid(aranged_ages, aranged_salaries)
#meshgrid converts to 2D of same size as arranged_ages and arranged_salaries are of diff sizes
```

```python
plt.contourf(X1, X2, 
             classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
             #converting the 2D to numpy array which is Y and transforming it and making it same size as X1 see pic in Day 7
             alpha = 0.7 , cmap = ListedColormap(('orange', 'blue')))

##Plotting Test set:
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)): #unique returns 0 and 1 and it is not of numpy array form so using enumerate
    plt.scatter(X_set[y_set == j, 0], 
                X_set[y_set == j, 1],
                c = ListedColormap(('red', 'white'))(j), 
                label =("Class", j))
plt.title('K Nearest Neighbors (Test set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
